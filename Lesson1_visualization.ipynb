## Data Visualization

Using the same axes :

It does not work for different scales and get little clumsy

Using axes :

It brings little clarity but we need to specify the coordinates of axes

Using Subplots

Brings clarity also we don't have to deal with coodinates of axes

```python
import pandas as pd
```

```python
data = pd.read_csv("dataset/adult.csv")
```

```python
data.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data['age'],data['capital.loss'],'--r')
plt.plot(data['age'],data['hours.per.week'],'|b')
plt.xlabel('Age')
```

```python
plt.axes([0.05,0.05,0.425,0.9])
plt.plot(data['age'],data['capital.loss'],'--r')
plt.xlabel('Age')
plt.title('loss vs age')

plt.axes([0.9,0.05,0.425,0.9])
plt.plot(data['age'],data['hours.per.week'],'.b')
plt.xlabel('Age')
plt.title('hours_of_wrk vs age')
```

```python
plt.subplot(1,2,1)
plt.plot(data['age'],data['capital.loss'],'--r')
plt.xlabel('Age')
plt.title('loss vs age')

plt.subplot(1,2,2)
plt.plot(data['age'],data['hours.per.week'],'.b')
plt.xlabel('Age')
plt.title('hours_of_wrk vs age')
```
