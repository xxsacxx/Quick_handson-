## Using wordnet

```python
from nltk.corpus import wordnet 
  
 
syns = wordnet.synsets("live") 
  
# An example of a synset: 
print(syns[0].name()) 
  
# Just the word: 
print(syns[0].lemmas()[0].name()) 
  
# Definition of that first synset: 
print(syns[0].definition()) 
  
# Examples of the word in use in sentences: 
print(syns[0].examples()) 
```

```python
def analogy(x1, x2, y1):
    result = model.most_similar(positive=[y1, x2], negative=[x1])
    print("As "+ x1 + " is related to " + x2 + " , similarly " + y1 +" is related to " + result[0][0])
    
```

```python
def display_pca_scatterplot(model, words=None, sample=0):
    if words == None:
        if sample > 0:
            words = np.random.choice(list(model.vocab.keys()), sample)
        else:
            words = [ word for word in model.vocab ]
        
    word_vectors = np.array([model[w] for w in words])

    twodim = PCA().fit_transform(word_vectors)[:,:2]
    
    plt.figure(figsize=(6,6))
    plt.scatter(twodim[:,0], twodim[:,1], edgecolors='k', c='r')
    for word, (x,y) in zip(words, twodim):
        plt.text(x+0.05, y+0.05, word)
```

```python
display_pca_scatterplot(model,['tywin','cersei','catelyn','cersei','daenerys','theon','tyrion','drogo','lannister','targaryen'])
```

```python
display_pca_scatterplot(model,['sachin','dhoni','dravid','yuvraj','narendra','modi','ranbir','shahrukh','homework', 'assignment', 'problem', 'exam', 'test', 'class',
                         'school', 'college', 'university', 'institute','maths','english','history'])
```
